```python
#DATA
import numpy as np

# Input features (Hours Studied)
X = np.array([1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0,
              5.5, 6.0, 6.5, 7.0, 7.5, 8.0, 8.5, 9.0, 9.5, 10.0])

# Output values (Exam Scores)
y = np.array([35, 45, 50, 55, 60, 62, 65, 70,
              75, 78, 80, 85, 88, 92, 95, 97, 99, 100])
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y,marker='*')
plt.xlabel("hours put in")
plt.ylabel("Marks")
```

```python
def Model(X,w,b):
  y_cap = np.dot(X,w) +b
  return y_cap


#test
y_cap = Model(X,-4,-29)

#grap
plt.plot(X,y_cap,color='r')
plt.scatter(X,y,marker="*")
plt.xlabel("hours put in")
plt.ylabel("Marks")
```

```python
#cost function
def cost_fun(y_cap,y):
  m = len(y)
  cost = np.subtract(y_cap,y)
  cost = cost**2
  cost = np.sum(cost)
  cost /= 2*m
  return cost

#test
y_cap = Model(X,-1,0)
cost = cost_fun(y_cap,y)
print(f"the cost for the choosen w and b is: {cost}")
```

```python
#gradient descnt
def gradient_des(X,y,alfa,itr,w,b):
    m = len(X)
    for c in range(itr) :
      y_cap = np.dot(X,w) + b
      w_in = np.dot(np.subtract(y_cap,y),X)
      b_in = np.sum(np.subtract(y_cap,y))
      w -= (alfa/m)*w_in
      b -= (alfa/m)*b_in
    return w,b
#test
w_new,b_new = gradient_des(X,y,0.01,10000,100,100)
print(f"w: {w_new}, b: {b_new}")
```

```python
y_c = Model(X,w_new,b_new)

#grap
plt.plot(X,y_c,color='r')
plt.scatter(X,y,marker="*")
plt.xlabel("hours put in")
plt.ylabel("Marks")
```

```python
cost_fun(y_c,y)
```
